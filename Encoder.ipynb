```python
def tokenize_pdf(string):
    lst_vectors = []
    for sentence in string:
        tokenizer.encode
        input_ids = tf.constant(tokenizer.encode(sentence,padding=True,truncation=True))[None, :]  # Batch size 1
        outputs = model(input_ids)
        last_hidden_states = outputs[0]
        lst_vectors.append(np.mean(last_hidden_states[0],axis=0))
    return lst_vectors
```
